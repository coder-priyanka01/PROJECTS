```python
import numpy as np
import matplotlib.pyplot as plt 
import pandas as pd

df = pd.read_csv('sales_data.csv')
df.head()
```

```python
print("\n first 5 rows of the data: \n", df.head())
print("\n last 5 rows of the data: \n", df.tail())
```

```python
print("\n shape of the data: ", df.shape)
print("\n data types and non-null counts: \n", df.info())
```

```python
print("\n total missing values in each column: \n", df.isnull().sum())
print("\n missing values in the column :")
print((df.isnull().sum() / len(df)) * 100)
```

```python
print("\n statistics of the data: \n", df.describe())

numeric_columns = df.select_dtypes(include=[np.number]).columns.tolist()
print("\n numeric columns: ", numeric_columns)    

categorical_columns = df.select_dtypes(include=['object']).columns.tolist()
print("\n categorical columns: ",categorical_columns)
```

```python
print("\n correlation matrix of the data: \n", df.corr(numeric_only=True))

# histograam of salary
plt.figure(figsize=(10, 6))
df['Salary'].hist(bins=10, color='lightblue', edgecolor='black')
plt.title('histogram of Salary')
plt.xlabel('Salary')
plt.ylabel('count')
plt.grid(True)
plt.tight_layout()
plt.show()

# histogram of age
plt.figure(figsize=(10, 6)) 
df['Age'].hist(bins=10, color='lightgreen', edgecolor='black')
plt.title('histogram of Age')
plt.xlabel('Age')
plt.ylabel('count')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# boxplot of salary
plt.figure(figsize=(10, 6))
df.boxplot(column='Salary', boxprops=dict(color='darkblue'))
plt.title('Boxplot of Salary')
plt.ylabel('Salary')
plt.grid(True)  
plt.tight_layout()
plt.show()

# boxplot of experience
plt.figure(figsize=(10, 6))
df.boxplot(column='Experience', boxprops=dict(color='orange'))
plt.title('Boxplot of Experience')
plt.ylabel('Experience')
plt.grid(True)
plt.tight_layout()
plt.show()
```
